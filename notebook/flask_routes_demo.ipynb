```python
from pathlib import Path
import requests

ROOT = Path.cwd()
if not (ROOT / "flask-app").exists():
    ROOT = ROOT.parent

BASE_URL = "http://127.0.0.1:5000"
print(ROOT)
```

```python
from PIL import Image

img_path = ROOT / "flask-app" / "sample_test.png"
Image.new("RGB", (80, 80), "red").save(img_path)
print(img_path.exists(), img_path)
```

```python
with open(img_path, "rb") as f:
    response = requests.post(
        f"{BASE_URL}/convert",
        files={"image": ("sample_test.png", f, "image/png")},
        data={"output_type": "JPEG"},
    )

print(response.status_code)
print(response.headers.get("Content-Type"))

out_file = ROOT / "flask-app" / "converted_from_notebook.jpeg"
out_file.write_bytes(response.content)
print(out_file.exists(), out_file.stat().st_size)
```

```python
payload = {
    "goal": "fat loss",
    "experience": "intermediate",
    "days_per_week": 3,
    "equipment": "dumbbells"
}

response = requests.post(f"{BASE_URL}/generate-workout-plan", json=payload)
print(response.status_code)

data = response.json()
print(data["message"])
print(data["input"])
print(data["exports"])
print(type(data["plan"]))
```
